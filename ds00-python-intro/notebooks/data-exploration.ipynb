```python
tmax_monthly = np.mean(tmax, axis=0)
tmax_mean = tmax.mean()
tmax_std = tmax.std()
tmax_monthly_std = tmax.std(axis=0)
tmax_upper = tmax_monthly + tmax_monthly_std
tmax_lower = tmax_monthly - tmax_monthly_std

ifg, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.plot(month, tmax_monthly, label='mean')
ax.fill_between(month, tmax_lower, tmax_upper, alpha=0.3, label='standard deviation')
ax.set_xlabel('Month')
ax.set_ylabel('Average Maximum Temperature [$^\circ$F]')
ax.axhline(y=tmax_mean, ls='--', color='k')
plt.legend()
```
